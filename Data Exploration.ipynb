```python
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
```

```python
dataset = pd.read_csv('dataset/noncompliant_customer_review_dataset.tsv', error_bad_lines=False,delimiter="\t",index_col=False)
```

```python
dataset.head()
```

```python
dataset.tail()
```

```python
dataset.sample(5)
```

```python
dataset['WORD_COUNT'] = dataset['REVIEW_TEXT'].apply(lambda x: len(str(x).split(" ")))
dataset[['REVIEW_TEXT','WORD_COUNT']].head()
```

```python
dataset['CHAR_COUNT'] = dataset['REVIEW_TEXT'].str.len() ## this also includes spaces
dataset[['REVIEW_TEXT','CHAR_COUNT']].head()
```

```python
def avg_word(sentence):
    words = sentence.split()
    return (sum(len(word) for word in words)/len(words))

dataset['AVG_WORD'] = dataset['REVIEW_TEXT'].apply(lambda x: avg_word(x))
dataset[['REVIEW_TEXT','AVG_WORD']].head()
```

```python
stop = stopwords.words('english')

dataset['stopwords'] = dataset['REVIEW_TEXT'].apply(lambda x: len([x for x in x.split() if x in stop]))
dataset[['REVIEW_TEXT','stopwords']].head()
```

```python
dataset['numerics'] = dataset['REVIEW_TEXT'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
dataset[['REVIEW_TEXT','numerics']].max()
```
